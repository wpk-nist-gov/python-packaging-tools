```python
import analphipy
import pandas as pd
```

```python
data = pd.read_csv("../examples/data/params.csv")
data.head()
```

```python
p = analphipy.potential.LennardJonesNM(n=10, m=4)
nf = p.to_nf()

nf.sig(beta=1 / 0.5)
```

```python
def calculate_effective_density(n, m, temp, dens):
    p = analphipy.potential.LennardJonesNM(n=n, m=m)
    nf = p.to_nf()

    sig = nf.sig(beta=1 / temp)

    return dens * sig**3
```

```python
dens_eff = []
for _, g in data.iterrows():
    dens_eff.append(calculate_effective_density(n=g.n, m=g.m, temp=g.temp, dens=g.dens))

data_eff = data.assign(dens_eff=dens_eff)
```
